```python
import pandas as pd
marketing_data = pd.read_csv('marketing_campaign.csv', sep="\t")
mnt = marketing_data.filter(regex='^Mnt.*')
mnt = pd.concat([marketing_data['ID'], mnt], axis=1)
marketing_data
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

for column in mnt.columns.drop('ID'):
    fig, ax = plt.subplots(figsize=(5, 5))
    sns.boxplot(data=marketing_data, x='Education', y=column, ax=ax)
    ax.set_title(f"{column} and Education")
    plt.show()
```
