First, we will do a little recap of web scraping. <br>

```python
import nltk
import requests

request = requests.get("https://www.gutenberg.org/cache/epub/8001/pg8001.html")
from bs4 import BeautifulSoup as bs

soup = bs(request.text, "html.parser")
import re
clean_content = re.sub(r'[\r|\n]+', '\n', soup.get_text())
print(clean_content)
```

## word tokenization

```python
nltk.download('punkt')
```

```python
contents_cleaned = nltk.word_tokenize(clean_content)
contents = list(filter(None, [(re.sub(r'[^a-zA-Z]', '', a)).lower() for a in contents_cleaned]))
```

```python
from collections import Counter
c = Counter(contents)
c.most_common(10)
```

```python
nltk.download('stopwords')
stopwords = nltk.corpus.stopwords.words('english')
contents_filtered = (word for word in contents  if word not in stopwords)
c = Counter(contents_filtered)
c.most_common(10)
```
